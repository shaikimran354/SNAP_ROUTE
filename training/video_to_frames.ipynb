```python
import os
import cv2
import numpy as np
import albumentations as A

# Path to the main directory where location folders exist
main_directory = "/content/drive/MyDrive/mini project"

# Define augmentations
transform = A.Compose([
    # Geometric Transformations
    A.Rotate(limit=30, p=1.0),  # Rotate ±30 degrees
    A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.1, rotate_limit=10, p=1.0),  # Shift, Scale, Rotate
    A.Perspective(scale=(0.05, 0.1), p=1.0),  # Perspective Transform

    # Color & Lighting Adjustments
    A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=1.0),
    A.RandomGamma(gamma_limit=(80, 120), p=1.0),
    A.HueSaturationValue(hue_shift_limit=20, sat_shift_limit=30, val_shift_limit=20, p=1.0),

    # Blurring & Noise
    A.GaussianBlur(blur_limit=5, p=1.0),
    A.MotionBlur(blur_limit=5, p=1.0),
    A.GaussNoise(var_limit=(10, 50), p=1.0),  # Fixed GaussNoise

    # Occlusion-Based Techniques
    A.CoarseDropout(max_holes=8, max_height=16, max_width=16, p=1.0),  # Fixed Cutout (Now CoarseDropout)
    A.GridDropout(ratio=0.2, p=1.0),  # GridMask

    # Weather & Environmental Simulation
    A.RandomFog(p=1.0),
    A.RandomShadow(shadow_roi=(0, 0, 1, 1), p=1.0)
])

# Loop through each location folder
for location in os.listdir(main_directory):
    location_path = os.path.join(main_directory, location)

    if os.path.isdir(location_path):  # Ensure it's a folder
        print(f"\nProcessing location: {location}")

        # Create an 'augmented' subfolder in the location folder
        augmented_folder = os.path.join(location_path, "augmented")
        if not os.path.exists(augmented_folder):
            os.makedirs(augmented_folder)

        # Get the list of already augmented images
        existing_augmented_files = set(os.listdir(augmented_folder))

        # Loop through all image files in the location folder
        for image_file in os.listdir(location_path):
            if image_file.endswith(('.jpg', '.jpeg', '.png')):  # Check for image formats
                aug_filename = f"aug_{image_file}"

                # Skip if already augmented
                if aug_filename in existing_augmented_files:
                    continue

                image_path = os.path.join(location_path, image_file)

                # Read the image
                image = cv2.imread(image_path)
                if image is None:
                    continue  # Skip if image cannot be read

                # Convert image to RGB (Albumentations uses RGB)
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

                # Apply augmentation
                augmented = transform(image=image)['image']

                # Convert back to BGR and save
                augmented = cv2.cvtColor(augmented, cv2.COLOR_RGB2BGR)
                cv2.imwrite(os.path.join(augmented_folder, aug_filename), augmented)

        print(f"  → Augmented images saved in: {augmented_folder}")

print("\n✅ Data Augmentation Completed!")
```

```python
import os
import tensorflow as tf
import cv2
import numpy as np

# Path to the main directory where location folders exist
main_directory = "/content/drive/MyDrive/mini project"

# Create a TensorFlow data augmentation pipeline
def get_augmentation_pipeline():
    return tf.keras.Sequential([
        # Geometric Transformations
        tf.keras.layers.RandomRotation(factor=0.1),  # Rotate ±10%
        tf.keras.layers.RandomTranslation(height_factor=0.1, width_factor=0.1),  # Shift
        tf.keras.layers.RandomZoom(height_factor=0.2, width_factor=0.2),  # Zoom

        # Color & Lighting Adjustments
        tf.keras.layers.RandomBrightness(0.2),  # Adjust brightness
        tf.keras.layers.RandomContrast(0.2),  # Adjust contrast

        # Blurring & Noise
        tf.keras.layers.GaussianNoise(stddev=0.05),  # Gaussian noise

        # Convert images back to uint8
        tf.keras.layers.Lambda(lambda x: tf.clip_by_value(x, 0, 255))
    ])

# Function to manually apply Gaussian blur and shadow
def apply_extra_augmentations(image):
    # Apply Gaussian Blur
    image = cv2.GaussianBlur(image, (5, 5), 0)

    # Create a random shadow effect
    h, w, _ = image.shape
    mask = np.zeros((h, w), dtype=np.uint8)
    x1, y1 = np.random.randint(0, w//2), np.random.randint(0, h//2)
    x2, y2 = np.random.randint(w//2, w), np.random.randint(h//2, h)
    cv2.rectangle(mask, (x1, y1), (x2, y2), 255, -1)
    shadow = cv2.GaussianBlur(mask, (21, 21), 0) / 255.0
    shadow = np.expand_dims(shadow, axis=-1)
    image = (image * (1 - shadow * 0.5)).astype(np.uint8)  # Darken shadow area

    return image

# Loop through each location folder
for location in os.listdir(main_directory):
    location_path = os.path.join(main_directory, location)

    if os.path.isdir(location_path):  # Ensure it's a folder
        print(f"\nProcessing location: {location}")

        # Create an 'augmented' subfolder in the location folder
        augmented_folder = os.path.join(location_path, "augmented")
        os.makedirs(augmented_folder, exist_ok=True)

        # Get the list of already augmented images
        existing_augmented_files = set(os.listdir(augmented_folder))

        # Augmentation Pipeline
        augmentor = get_augmentation_pipeline()

        # Loop through all image files in the location folder
        for image_file in os.listdir(location_path):
            if image_file.endswith(('.jpg', '.jpeg', '.png')):  # Check for image formats
                image_path = os.path.join(location_path, image_file)
                image = cv2.imread(image_path)

                if image is None:
                    continue  # Skip if image cannot be read

                # Convert image to RGB (TensorFlow uses RGB format)
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                image = tf.convert_to_tensor(image, dtype=tf.float32)

                # Generate 2 augmented images per original
                for i in range(2):
                    aug_image = augmentor(tf.expand_dims(image, axis=0))
                    aug_image = tf.squeeze(aug_image).numpy().astype(np.uint8)

                    # Apply additional augmentations (Blur & Shadow)
                    aug_image = apply_extra_augmentations(aug_image)

                    # Convert back to BGR and save
                    aug_image = cv2.cvtColor(aug_image, cv2.COLOR_RGB2BGR)
                    aug_filename = f"aug_{i}_{image_file}"
                    cv2.imwrite(os.path.join(augmented_folder, aug_filename), aug_image)

        print(f"  → Augmented images saved in: {augmented_folder}")

print("\n✅ Data Augmentation Completed!")
```
